```python
from collections import Counter
name = "Nallaganti Sharath Babu"

name = name.replace(" ", "").lower()

print(Counter(name))

freq = Counter(name)
most_char,most_count = freq.most_common(1)[0]

print("most common character:",most_char)
print("Count",most_count)
```
