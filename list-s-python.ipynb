# list

A list is a built-in data structure in Python used to store multiple values in a single variable.

Characteristics of a list:

Ordered (items have index positions)

Mutable (can be changed)

Allows duplicate values

Can store different data types

```python
my_list = ["sharath", 14.6, 17151]
```

# Common List Operations

* Creating a list

* Accessing elements

* Updating elements

* Adding elements

* Removing elements

* Searching elements

* Sorting

* Copying

* Iterating

```python
# Creating a list
my_list = ["sharath", 14.6, 17151, 10, 20]

print("Original list:", my_list)

# Accessing elements
print("First element:", my_list[0])
print("Last element:", my_list[-1])

# Updating elements
my_list[1] = 15.0

# Adding elements
my_list.append("Python")
my_list.insert(2, "Code")
my_list.extend([30, 40])

# Removing elements
my_list.remove(10)
my_list.pop()
del my_list[1]

# Searching elements
print("Index of 20:", my_list.index(20))
print("Count of 20:", my_list.count(20))

# Length of list
print("Length of list:", len(my_list))

# Sorting
num_list = [5, 2, 9, 1]
num_list.sort()
print("Sorted list:", num_list)

# Reversing
num_list.reverse()
print("Reversed list:", num_list)

# Copying a list
copy_list = my_list.copy()

# Membership test
print("Is 'Python' in list?", "Python" in my_list)

# Iterating through list
print("Iterating list:")
for item in my_list:
    print(item)

# Slicing
print("Sliced list:", my_list[1:4])

# List comprehension
squares = [x * x for x in range(1, 6)]
print("Squares:", squares)

# Final output
print("Final modified list:", my_list)
print("Copied list:", copy_list)
```
